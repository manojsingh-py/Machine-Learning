```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('../datasets/train.csv')[['Age','Pclass','Fare','Survived']]
```

```python
df.head()
```

```python
df.isnull().mean() * 100
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.head()
```

```python
# knn_imputer = KNNImputer(n_neighbors=3)
# knn_imputer = KNNImputer(n_neighbors=3, weights='uniform', metric='nan_euclidean')
knn_imputer = KNNImputer(n_neighbors=3, weights='distance', metric='nan_euclidean')
X_train_trf = knn_imputer.fit_transform(X_train)
X_test_trf = knn_imputer.transform(X_test)
```

```python
X_train_trf[:5]
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf,y_train)

y_pred = lr.predict(X_test_trf)

accuracy_score(y_test,y_pred)
```

```python
# Comparision with Simple Imputer --> mean
si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
lr2 = LogisticRegression()
lr2.fit(X_train_trf2, y_train)
y_pred2 = lr2.predict(X_test_trf2)
accuracy_score(y_test, y_pred2)
```
