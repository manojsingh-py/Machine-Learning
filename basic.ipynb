````python
# Basic Decorator with *args and **kwargs

def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before function call")

        result = func(*args, **kwargs)  # pass all arguments

        print("After function call")
        return result

    return wrapper
````

````python
@my_decorator
def greet(*args, **kwargs):
    print(f"Hello {args[0]}, age {args[1]}, city {kwargs['city']}, name {kwargs['name']}")

greet("Manoj", 25, city="Pune", name="Manoj")
````

````python
# Real-World Example: Logging Decorator

def log_function(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}")
        print(f"Args: {args}, Kwargs: {kwargs}")

        result = func(*args, **kwargs)

        print(f"{func.__name__} returned {result}")
        return result

    return wrapper
````

````python
@log_function
def multiply(a, b):
    return a * b

multiply(3, 4)
````

````python
# Decorator with Return Value Handling

def double_result(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result * 2
    return wrapper
````

````python
@double_result
def get_number():
    return 5

print(get_number())  # Output: 10
````

````python
# Structure of Decorator with Arguments

def decorator_with_args(param):
    def actual_decorator(func):
        def wrapper(*args, **kwargs):
            # logic using param
            return func(*args, **kwargs)
        return wrapper
    return actual_decorator

# Outer function → takes decorator arguments (param)
# Middle function → takes the function (func)
# Inner function → wraps execution (wrapper)
````

````python
# Example 1: Simple Parameterized Decorator

def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
````

````python
@repeat(3)
def say_hello():
    print("Hello!")

say_hello()
````

````python
def log(prefix):
    def decorator(func):
        def wrapper(*args, **kwargs):
            print(f"{prefix} Calling {func.__name__}")
            return func(*args, **kwargs)
        return wrapper
    return decorator
````

````python
@log("DEBUG")
def add(a, b):
    return a + b

print(add(2, 3))
````

````python
# Example 3: Access Control (Real-World Pattern)

def require_role(role):
    def decorator(func):
        def wrapper(*args, **kwargs):
            user_role = kwargs.get("role", "guest")

            if user_role != role:
                print("Access denied")
                return None

            return func(*args, **kwargs)
        return wrapper
    return decorator
````

````python
@require_role("admin")
def delete_user(user_id, role=None):
    print(f"User {user_id} deleted")

delete_user(101, role="admin")   # Allowed
delete_user(102, role="guest")   # Denied
````

````python
class A:
    def __init__(self, a):
        self.a = a

    def a_has(self):
        print(self.a)


class B:
    def __init__(self, b):
        self.b = b

    def b_has(self):
        print(self.b)


class C(A, B):
    def __init__(self, x, y, z):
        self.a = x
        self.b = y
        self.c = z

    def c_has(self):
        self.a_has()
        self.b_has()
        print(self.c)


print(C.__mro__)
c = C(1, 2, 3)
c.c_has()
````

````python
# __init__ vs __new__ in Inheritance

class A:
    def __new__(cls):
        print("A new")
        return super().__new__(cls)

    def __init__(self):
        print("A init")

class B(A):
    def __new__(cls):
        print("B new")
        return super().__new__(cls)

    def __init__(self):
        print("B init")
        super().__init__()

class C(A):
    def __new__(cls):
        print("C new")
        return super().__new__(cls)

    def __init__(self):
        print("C init")
        super().__init__()

class D(B, C):
    def __new__(cls):
        print("D new")
        return super().__new__(cls)

    def __init__(self):
        print("D init")
        super().__init__()

D()

# 🔹 Key Insight
#
# ✔️ Both __new__ and __init__ follow MRO
# ✔️ But:
#
# __new__ → creates object (top-down chain)
# __init__ → initializes object (also via super() chain)
````

## 🔹 `__new__` in Inheritance & MRO

When you combine **inheritance**, **MRO**, and `__new__`, the key idea is:

> 🔥 `__new__` follows the **same MRO chain as methods**, just like `super()`

---

# 🔹 1. Quick Refresher

* `__new__(cls, ...)` → **creates the object**
* Runs **before** `__init__`
* Must return an instance → usually via `super().__new__(cls)`

---

# 🔹 2. `__new__` Follows MRO

Let’s reuse a diamond-style hierarchy:

```python
class A:
    def __new__(cls):
        print("A __new__")
        return super().__new__(cls)

class B(A):
    def __new__(cls):
        print("B __new__")
        return super().__new__(cls)

class C(A):
    def __new__(cls):
        print("C __new__")
        return super().__new__(cls)

class D(B, C):
    def __new__(cls):
        print("D __new__")
        return super().__new__(cls)

d = D()
```

---

## 🔹 MRO

```python
print(D.__mro__)
```

👉

```
(D, B, C, A, object)
```

---

## 🔹 Execution Order (`__new__`)

Output:

```
D __new__
B __new__
C __new__
A __new__
```

✔️ Exactly follows MRO:
**D → B → C → A → object**

---

# 🔹 3. What’s Happening Internally?

Each class does:

```python
return super().__new__(cls)
```

👉 This means:

* “Pass control to the **next class in MRO**”
* Not necessarily the direct parent

---

# 🔹 4. Important Rule ⚠️

👉 If ANY class in the chain **does NOT call `super().__new__()`**, the chain breaks.

### ❌ Example (Breaking MRO chain):

```python
class B(A):
    def __new__(cls):
        print("B __new__")
        return object.__new__(cls)  # ❌ breaks chain
```

👉 Now:

* `C.__new__` and `A.__new__` will **never run**

## 🔹 The Core Idea

In multiple inheritance, Python uses **MRO (C3 linearization)** to create a **single, consistent call chain**:

```
D → B → C → A → object
```

`super()` means:

> “Call the **next method in the MRO**, not my direct parent.”

That’s what lets all classes in the hierarchy **participate exactly once**.

---

## 🔹 What Goes Wrong Without `super()`

### ❌ Problem 1: Skipping classes

```python
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")
        A.show(self)   # ❌ direct call

class C(A):
    def show(self):
        print("C")
        super().show()

class D(B, C):
    def show(self):
        print("D")
        super().show()

D().show()
```

👉 Output:

```
D
B
A
```

⚠️ `C.show()` is **never called**
→ because `B` bypassed MRO by calling `A` directly

---

### ❌ Problem 2: Duplicate calls

```python
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")
        A.show(self)

class C(A):
    def show(self):
        print("C")
        A.show(self)

class D(B, C):
    def show(self):
        print("D")
        B.show(self)
        C.show(self)

D().show()
```

👉 Output:

```
D
B
A
C
A
```

⚠️ `A` runs **twice** → duplicate execution
→ this is the classic **diamond problem**

---

## 🔹 Correct Approach with `super()`

```python
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")
        super().show()

class C(A):
    def show(self):
        print("C")
        super().show()

class D(B, C):
    def show(self):
        print("D")
        super().show()

D().show()
```

👉 Output:

```
D
B
C
A
```

✔️ Every class runs **once and in correct order**

---

## 🔹 Why `super()` Fixes Everything

### ✅ 1. Respects MRO

It follows:

```
D → B → C → A
```

### ✅ 2. Avoids duplication

Each class appears **only once** in MRO

### ✅ 3. Enables cooperative inheritance

Every class says:

> “Do my work, then pass control forward”

---

## 🔹 Mental Model

Think of `super()` like a **relay race baton** 🏃

* Each class runs its part
* Then passes control to the **next runner (MRO)**
* No skipping, no repeating

---

## 🔹 Important Rule ⚠️

👉 Every class in the chain must use:

```python
super().method()
```

If even **one class breaks the chain**, everything after it is affected.

---

## 🔥 Interview Answer (Strong)

> In multiple inheritance, `super()` is required to ensure that method calls follow the MRO (C3 linearization). It prevents duplicate executions and ensures all classes in the hierarchy are called exactly once in the correct order.

---

## 🔹 When is `super()` NOT critical?

* Single inheritance → direct parent call *might* work
* Multiple inheritance → **always use `super()`**
