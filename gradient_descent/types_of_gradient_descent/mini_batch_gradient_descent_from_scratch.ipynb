```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X, y = load_diabetes(return_X_y=True)
```

```python
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.coef_)
print(lr.intercept_)
```

```python
y_pred = lr.predict(X_test)
print(r2_score(y_test, y_pred))
```

```python
import random


class MiniBatchGradientDescent:
    def __init__(self, learning_rate=0.01, batch_size=32, epochs=100):
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        self.epochs = epochs
        self.coef = None
        self.intercept = None

    def fit(self, X, y):
        # Init coefficients and intercept
        self.intercept = 0
        self.coef = np.ones(X.shape[1])
        for i in range(self.epochs):
            for j in range(int(X.shape[0] / self.batch_size)):
                idx = np.random.randint(0, X.shape[0], int(X.shape[0] / self.batch_size))
                # idx = random.sample(idx, int(X.shape[0] / self.batch_size))
                y_hat = np.dot(X[idx], self.coef) + self.intercept

                intercept_der =  -2 * np.mean(y[idx] - y_hat)
                self.intercept = self.intercept - (self.learning_rate * intercept_der)

                coef = -2 * np.dot((y[idx] - y_hat), X[idx])
                self.coef = self.coef - (self.learning_rate * coef)

        print(f"Final Coefficients: {self.coef}, Final Intercept: {self.intercept}")

    def predict(self, X):
        return np.dot(X, self.coef) + self.intercept

    def score(self, X, y):
        y_pred = self.predict(X)
        return r2_score(y, y_pred)
```

```python
mbgd = MiniBatchGradientDescent(learning_rate=0.01, batch_size=40, epochs=100)
mbgd.fit(X_train, y_train)
```

```python
print(f"Coefficients: {mbgd.coef}")
print(f"Intercept: {mbgd.intercept}")

print(f"R^2 Score on Training Data: {mbgd.score(X_train, y_train)}")
```
