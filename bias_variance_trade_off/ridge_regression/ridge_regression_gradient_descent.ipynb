```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
from statsmodels.tsa.forecasting import theta
```

```python
data=load_diabetes()
```

```python
X=data.data
y=data.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=45)
```

```python
from sklearn.linear_model import SGDRegressor

sgd_model = SGDRegressor(penalty='l2', alpha=1.0, max_iter=1000, tol=1e-3)
sgd_model.fit(X_train, y_train)
print('Coef', sgd_model.coef_)
print('Intercept', sgd_model.intercept_)
```

```python
y_pred = sgd_model.predict(X_test)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(np.mean((y_test - y_pred)**2))
print('R2 Score:', r2)
print('RMSE:', rmse)
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)
print('Coef', ridge_model.coef_)
print('Intercept', ridge_model.intercept_)
```

```python
y_pred1=ridge_model.predict(X_test)
r2 = r2_score(y_test, y_pred1)
rmse = np.sqrt(np.mean((y_test - y_pred1)**2))
print('R2 Score:', r2)
print('RMSE:', rmse)
```

```python
reg = SGDRegressor(penalty='l2',max_iter=500,eta0=0.1,learning_rate='constant',alpha=0.001)

reg.fit(X_train, y_train)
print('Coef', reg.coef_)
print('Intercept', reg.intercept_)
```

```python
y_pred2 = reg.predict(X_test)
r2 = r2_score(y_test, y_pred2)
rmse = np.sqrt(np.mean((y_test - y_pred2)**2))
print('R2 Score:', r2)
print('RMSE:', rmse)
```

```python
class CustomRidge:
    def __init__(self, alpha=1.0, learning_rate=0.01, epochs=1000):
        self.alpha = alpha
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.coef_ = None
        self.intercept_ = None

    def fit(self, X, y):

        self.coef_ = np.ones(X_train.shape[1])
        self.intercept_ = 0
        X = np.insert(X,0,1,axis=1)
        theta = np.insert(self.coef_,0,self.intercept_)

        for _ in range(self.epochs):
            theta_der = np.dot(X.T, X).dot(theta) - np.dot(X.T, y) + self.alpha * self.learning_rate
            theta = theta - self.learning_rate * theta_der

            self.coef_ = theta[1:]
            self.intercept_ = theta[0]

    def predict(self, X):
        return np.dot(X, self.coef_) + self.intercept_
```

```python
custom_ridge = CustomRidge(alpha=0.001, learning_rate=0.001, epochs=100)
custom_ridge.fit(X_train, y_train)
print('Coef:', custom_ridge.coef_)
print('Intercept:', custom_ridge.intercept_)
y_pred_custom = custom_ridge.predict(X_test)
r2 = r2_score(y_test, y_pred_custom)
rmse = np.sqrt(np.mean((y_test - y_pred_custom)**2))
print('R2 Score:', r2)
print('RMSE:', rmse)
```
